```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

caminho='/content/drive/My Drive/API 2026 - Cargas perigosas/'
```

```python
df = pd.read_csv(caminho + "Dadostratados.csv")
```

```python
# Isolar a coluna "produto"
produtos = df['Produto']

# Contar a aparição de cada produto
contagem_produtos = produtos.value_counts()

# Listar os 10 produtos mais frequentes
print("Top 10 Produtos mais frequentes:")
display(contagem_produtos.head(10))

# To study a specific product, you can filter the DataFrame:
# For example, to see all entries for 'Aspartame':
# aspartame_df = df[df['Produto'] == 'Aspartame']
# display(aspartame_df.head())

# Or, to get unique products:
# unique_produtos = produtos.unique()
# print("\nNúmero de produtos únicos:", len(unique_produtos))
```

```python
# Get all unique products
unique_produtos = produtos.unique()

print(f"Número total de produtos únicos: {len(unique_produtos)}")
print("Todos os produtos únicos:")
display(unique_produtos)
```

```python
# Convert the unique products array to a DataFrame
df_produtos_unicos = pd.DataFrame(unique_produtos, columns=['Produto_Unico'])

# Define the output path in Google Drive
output_path = caminho + 'produtos_unicos.csv'

# Export to CSV
df_produtos_unicos.to_csv(output_path, index=False)

print(f"Lista de produtos únicos exportada para: {output_path}")
```

### Filtrando com a lista de produtos perigosos

```python
# Carregar a lista de produtos perigosos tratados
df_produtos_perigosos = pd.read_csv(caminho + 'produtos_perigosos_tratados.csv')

# Extrair a lista de nomes de produtos da coluna 'Produto_Unico'
# Assumindo que a coluna com os nomes dos produtos se chama 'Produto_Unico'
# ou a primeira coluna se o nome for desconhecido.
# Ajuste 'Produto_Unico' se o nome da coluna for diferente no CSV.
if 'Produto_Unico' in df_produtos_perigosos.columns:
    lista_produtos_para_filtrar = df_produtos_perigosos['Produto_Unico'].unique()
elif len(df_produtos_perigosos.columns) > 0:
    # Se não encontrar 'Produto_Unico', tenta a primeira coluna
    lista_produtos_para_filtrar = df_produtos_perigosos.iloc[:, 0].unique()
else:
    lista_produtos_para_filtrar = []
    print("Não foi possível encontrar a coluna de produtos ('Produto_Unico' ou primeira coluna) no arquivo 'produtos_perigosos_tratados.csv'.")

print(f"Número de produtos na lista de filtragem (produtos perigosos): {len(lista_produtos_para_filtrar)}")
```

```python
# Aplicar o filtro no DataFrame original (df)
df_filtered_perigosos = df[df['Produto'].isin(lista_produtos_para_filtrar)]

print(f"Número original de linhas: {len(df)}")
print(f"Número de linhas após o filtro de produtos perigosos: {len(df_filtered_perigosos)}")

display(df_filtered_perigosos.head())
```

```python
# Contar a aparição de cada produto na coluna 'Produto' do DataFrame filtrado
contagem_produtos_filtrados = df_filtered_perigosos['Produto'].value_counts()

print("Top 10 Produtos mais frequentes no DataFrame filtrado por produtos perigosos:")
display(contagem_produtos_filtrados.head(10))

print(f"Número total de produtos únicos no DataFrame filtrado: {len(contagem_produtos_filtrados)}")
```

### Salvando o DataFrame Filtrado

```python
# Definir o nome do novo arquivo CSV
nome_arquivo_saida = 'Dados Tratados - Filtrado.csv'

# Construir o caminho completo para o arquivo de saída
caminho_arquivo_saida = caminho + nome_arquivo_saida

# Salvar o DataFrame df_filtered_perigosos em CSV
df_filtered_perigosos.to_csv(caminho_arquivo_saida, index=False)

print(f"DataFrame filtrado salvo com sucesso em: {caminho_arquivo_saida}")
```
